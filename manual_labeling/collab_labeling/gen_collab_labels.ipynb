```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('./all_cleaned.csv')
df = df[['video_id','channel_id','rank','component','watch_video_id','bot_name','phase','homepage_level','videopage_level','topic','strategy','note']]
df = df[(df['topic']!='random')&(df['rank']<10)]
df.rename(columns={'stain':'on_list'},inplace=True)
df.head()
```

```python
df_homepage = df[
    (df['component']=='homepage')&
    (df['homepage_level'].isin([1,39,79]))
]
```

```python
df_videopage = df[(df['phase']=='videopage_experiment')]
df_videopage.head()
```

```python
df_all = pd.concat([df_homepage,df_videopage])
```

```python
label_sheet = pd.get_dummies(df_all[['channel_id','topic']].set_index('channel_id')).groupby('channel_id').agg('sum').apply(lambda x: x.apply(lambda y: 1 if y>0 else 0)).replace({0: pd.NA, 1: '?'})
label_sheet.index = label_sheet.reset_index()['channel_id'].apply(lambda x: f'https://www.youtube.com/channel/{x}')
label_sheet

```

```python
label_sheet.to_csv('master_collab_label.csv')
```

```python
labeled_df = pd.read_csv('../bot_runs_labeled_1-4.csv')
labeled_df['channel_id'] = labeled_df['channel_id'].apply(lambda x: f'https://www.youtube.com/channel/{x}')
labeled_df = labeled_df[['channel_id','topic','on_topic']]
labeled_df.drop_duplicates(inplace=True)
labeled_df.set_index(['channel_id','topic'],inplace=True)
labeled_df.head()
```

```python
max_idx_50_each_topic = max([label_sheet.reset_index()[~label_sheet.reset_index()[f'topic_{topic}'].isna()].index[:50][-1] for topic in [
    'alt-right',
    'antitheist',
    'politics-left',
    'politics-right'
]])
```

```python
label_sheet_done = pd.DataFrame()
for topic in labeled_df.reset_index()['topic'].unique():
    label_sheet_done[topic] = label_sheet.iloc[:max_idx_50_each_topic,:].join(labeled_df.loc[(slice(None),topic),:].droplevel(-1))['on_topic']
```
